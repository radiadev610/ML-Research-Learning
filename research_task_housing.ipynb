```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

# Step 1: Baseline Data Load Only
df_house = pd.read_csv(r'C:\Users\USER\Downloads\Housing.csv')
df = df_house.copy()
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

num_cols = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']
cat_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']

X_num = df[num_cols]
y = df['price']


# Step 2: Baseline One-Hot Encoding
X_cat_onehotencoding = pd.get_dummies(df[cat_cols], drop_first=False)

# Combine numerical and one-hot encoded variables
X_baseline = pd.concat([X_num, X_cat_onehotencoding], axis=1)

print("=== STEP 2: BASELINE ONE-HOT ENCODING RESULTS ===")
print(f"Baseline Data Table Shape: {X_baseline.shape}")
print(f"Total Columns Generated: {X_baseline.shape[1]}")

# Step 3 & 5: Enhanced Encoding & Controlled Experiment

X_cat_freq = df[cat_cols].copy()

for col in cat_cols:
    freq_map = df[col].value_counts(normalize=True)
    X_cat_freq[col] = X_cat_freq[col].map(freq_map)

X_enhanced = pd.concat([X_num, X_cat_freq], axis=1)

print("\nSTEP 3: ENHANCED FREQUENCY ENCODING RESULTS")
print(f"Enhanced Data Table Shape: {X_enhanced.shape}")
print(f"Total Columns Generated: {X_enhanced.shape[1]}")

model = RandomForestRegressor(random_state=42)

scores_baseline = cross_val_score(model, X_baseline, y, cv=5)
scores_enhanced = cross_val_score(model, X_enhanced, y, cv=5)

mean_baseline_r2 = scores_baseline.mean()
mean_enhanced_r2 = scores_enhanced.mean()

print("\nSTEP 5: CONTROLLED EXPERIMENT METRICS")
print(f"Baseline One-Hot Encoding Mean R2 Score: {mean_baseline_r2:.4f}")
print(f"Enhanced Frequency Encoding Mean R2 Score: {mean_enhanced_r2:.4f}")
```
