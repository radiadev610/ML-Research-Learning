```python
import pandas as pd
import numpy as np

sales_data = {
 'Region' : ['North','North','South','South','East','East','West','West'],
 'Product' : ['A','B','A','B','A','B','A','B'],
 'Units' : [120, 85, 95, 110, 60, 75, 130, 90],
 'Revenue' : [2400, 1700, 1900, 2200, 1200, 1500, 2600, 1800],
}
sales = pd.DataFrame(sales_data)
print(sales)
```

```python
print("TASK 7")
region_rev = sales.groupby('Region')['Revenue'].sum()
print("Total Revenue per Region:")
print(region_rev)

product_units = sales.groupby('Product')['Units'].mean()
print("\nAverage Units sold per Product:")
print(product_units)

pivot = sales.pivot_table(index='Region', columns='Product', values='Revenue', aggfunc='sum', margins=True)
print("\nPivot Table (Total Revenue):")
print(pivot)
```

```python
print("TASK 8")
sales['Revenue_per_Unit'] = (sales['Revenue'] / sales['Units']).round(2)
region_efficiency = sales.groupby('Region')['Revenue_per_Unit'].mean().sort_values(ascending=False)
print(region_efficiency)
print(f"The most efficient region is {region_efficiency.index[0]} with an average revenue per unit of {region_efficiency.iloc[0]}")
```

```python
print("TASK 9")
print("Raw Counts:")
print(sales['Product'].value_counts())
print("\nPercentages:")
print(sales['Product'].value_counts(normalize=True).round(4) * 100)
```
