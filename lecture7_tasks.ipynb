```python
import pandas as pd
import numpy as np

np.random.seed(10)
students = [f'Student_{i}' for i in range(1, 31)]
classes = np.random.choice(['A', 'B', 'C'], 30)
subjects = ['Math', 'Science', 'English', 'ICT']

rows = []
for s, c in zip(students, classes):
    for subj in subjects:
        rows.append({
            'Student': s,
            'Class': c,
            'Subject': subj,

            'Marks': np.random.randint(30, 100)
        })

df = pd.DataFrame(rows)
print(df.head(10))
print(f"Shape: {df.shape}")
```

```python
print("TASK 1")
class_avg = df.groupby('Class')['Marks'].mean().round(1)
print(class_avg)
highest_class = class_avg.idxmax()
print(f"The class with the highest average is Class {highest_class}\n")
```

```python
print("TASK 2")
subject_avg = df.groupby('Subject')['Marks'].mean().sort_values(ascending=False).round(1)
print(subject_avg)
print("\n")
```

```python
print("TASK 3")
student_avg = df.groupby('Student')['Marks'].mean().sort_values(ascending=False).round(1)
print(f"Best student: {student_avg.index[0]} with average {student_avg.iloc[0]}")
print(f"Worst student: {student_avg.index[-1]} with average {student_avg.iloc[-1]}\n")
```

```python
print("TASK 4")
df['Result'] = np.where(df['Marks'] >= 40, 'Pass', 'Fail')
ct = pd.crosstab(df['Class'], df['Result'])
print(ct)
```

```python
print("TASK 5")
subject_summary = df.groupby('Subject')['Marks'].agg(['min', 'max', 'count'])
print(subject_summary)
```

```python
print("TASK 6")
class_subject_avg = df.groupby(['Class', 'Subject'])['Marks'].mean().round(1)
print(class_subject_avg)
```
