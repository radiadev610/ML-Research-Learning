```python
import pandas as pd
employees = pd.DataFrame({
'EmpID': [101, 102, 103, 104, 105],
'Name': ['Asha', 'Rohan', 'Meera', 'Karan', 'Tina'],
'Department': ['HR', 'IT', 'IT', 'Finance', 'HR'],
})
salaries = pd.DataFrame({
'EmpID': [101, 102, 103, 106],
'Salary': [45000, 60000, 55000, 50000],
})

print("SECTION A\n")
print("QUESTION 1")
q1_merge = pd.merge(employees, salaries, on='EmpID') 
print(q1_merge)
print(f"Number of rows: {len(q1_merge)}\n")

print("QUESTION 2")
q2_merge = pd.merge(employees, salaries, on='EmpID', how='left')
print(q2_merge)
print("Missing salary value appears as: NaN\n")

print("QUESTION 3")
q3_merge = pd.merge(employees, salaries, on='EmpID', how='outer')
print(q3_merge)
print(f"Number of rows: {len(q3_merge)}")
print("EmpIDs appearing not in left merge: 106\n")
```

```python
import pandas as pd
info = pd.DataFrame({'Name': ['Asha', 'Rohan', 'Meera']}, index=[101, 102, 103])
ratings = pd.DataFrame({'Rating': [4.5, 3.8, 4.2]}, index=[101, 102, 103])
batch_march = pd.DataFrame({'Name': ['Asha', 'Rohan'], 'Sales': [120, 95]})
batch_april = pd.DataFrame({'Name': ['Meera', 'Karan'], 'Sales': [110, 130]})

print("SECTION B\n")
print(" QUESTION 4")
q4_join = info.join(ratings)
print(q4_join)
print("This works without specifying a key column because .join() automatically combines DataFrames using their index labels.\n") 

print(" QUESTION 5")
all_sales = pd.concat([batch_march, batch_april], ignore_index=True)
print(all_sales)
print("\n")

print("QUESTION 6")
batch_march['Month'] = 'March'
batch_april['Month'] = 'April'
all_sales = pd.concat([batch_march, batch_april], ignore_index=True) 
print(all_sales)
print("\n")

print("QUESTION 7")
q7_concat = pd.concat([info, ratings], axis=1)
print(q7_concat)
```

```python
import pandas as pd
wide_scores = pd.DataFrame({
'Name': ['Asha', 'Rohan', 'Meera'],
'Quiz1': [8, 6, 9],
'Quiz2': [7, 8, 10],
'Quiz3': [9, 7, 8],
})

print("SECTION C\n")
print("QUESTION 8")
long_scores = wide_scores.melt(id_vars='Name', value_vars=['Quiz1', 'Quiz2', 'Quiz3'], var_name='Quiz', value_name='Score') 
print(long_scores)
print("\n")

print("QUESTION 9")
back_to_wide = long_scores.pivot(index='Name', columns='Quiz', values='Score')
print(back_to_wide)
print("\n")

print("QUESTION 10")
student_avg = long_scores.groupby('Name')['Score'].mean()
print(student_avg)
print(f"The student with the highest average is {student_avg.idxmax()}\n")

print("QUESTION 11")
quiz_avg = long_scores.groupby('Quiz')['Score'].mean()
print(quiz_avg)
print(f"The quiz with the lowest average score is {quiz_avg.idxmin()}\n")
```

```python
import pandas as pd
records = pd.DataFrame({
'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
'Quarter': ['Q1', 'Q2', 'Q1', 'Q2', 'Q1', 'Q2'],
'Expense': [50000, 52000, 30000, 31000, 45000, 47000],
})

print("SECTION D\n")
print("QUESTION 12")
grouped_expenses = records.groupby(['Department', 'Quarter'])['Expense'].sum()
print(grouped_expenses)
print("\n")

print("QUESTION 13")
q2_hr_expense = grouped_expenses.loc[('HR', 'Q2')]
print(q2_hr_expense)
print("\n")

print("QUESTION 14")
flat_expenses = grouped_expenses.reset_index()
expense_pivot = flat_expenses.pivot(index='Department', columns='Quarter', values='Expense')
print(expense_pivot)
print("\n")
```

```python
import pandas as pd
students = pd.DataFrame({
'StudentID': [1, 2, 3, 4],
'Name': ['Asha', 'Rohan', 'Meera', 'Karan'],
'Section': ['A', 'A', 'B', 'B'],
})
scores_term1 = pd.DataFrame({
'StudentID': [1, 2, 3, 4],
'English': [70, 65, 80, 60],
'Maths': [85, 60, 75, 90],
})
scores_term2 = pd.DataFrame({
'StudentID': [1, 2, 3, 4],
'English': [75, 68, 82, 64],
'Maths': [88, 63, 78, 92],
})

print("SECTION E\n")

print("QUESTION 15")
scores_term1['Term'] = 'Term1'
scores_term2['Term'] = 'Term2'
all_scores_wide = pd.concat([scores_term1, scores_term2], ignore_index=True)
print(all_scores_wide)
print("\n")

print("QUESTION 16")
all_scores_long = all_scores_wide.melt(id_vars=['StudentID', 'Term'], value_vars=['English', 'Maths'], var_name='Subject', value_name='Score')
print(all_scores_long)
print("\n")

print("QUESTION 17")
merged_scores = pd.merge(all_scores_long, students, on='StudentID', how='left')
print(merged_scores)
print("\n")

print("QUESTION 18")
section_summary = merged_scores.pivot_table(index=['Section', 'Term'], columns='Subject', values='Score', aggfunc='mean').round(1)
print(section_summary)
print("\n")

print("QUESTION 19")
term_pivot = merged_scores.pivot_table(index=['StudentID', 'Name', 'Subject'], columns='Term', values='Score').reset_index()
print(all_scores_long)
print("\n")


print("QUESTION 20")
term_pivot['Improvement'] = term_pivot['Term2'] - term_pivot['Term1']
sorted_improvement = term_pivot.sort_values(by='Improvement', ascending=False)
print(sorted_improvement.head(1))
print("\n")

merged_scores.to_csv('student_scores_combined.csv', index=False)
print("FILE WRITTEN SUCCESSFULLY")
```
